```python
# Import necessary libraries
import matplotlib.pyplot as plt
import numpy as np

# Define the data from global_stats_squad.py as dictionaries for plotting

# Accuracy for each system
accuracy_squad = {
    'Baseline LLM': 0.48,
    'EDC GoG Pipeline': 0.62,
    'Integrated EDC GoG Refinement': 0.68,
}

# Query Answering (QA) Accuracy for Incomplete graph (100 examples)
qa_accuracy_squad = {
    'EDC Pipeline': 0.12,
    'GoG Pipeline': 0.53,
    'Integrated EDC GoG Refinement': 0.49
}

# Execution Time (in minutes) QA - 500 examples
execution_time_squad = {
    'EDC Graph Creation': 185,
    'QA GoG': 153,
    'QA Integrated': 246,
    'Accuracy Calculation': 24,
}

# Scalability (Processing Time in minutes for varying data sizes) (10% QA))
scalability_squad = {
    '5 Titles': [49, 42, 68, 7],
    '20 Titles': [185, 153, 246, 24],
    '50 Titles': [453, 378, 612, 57],
}

# Plotting the data

# Accuracy Plot
plt.figure(figsize=(10, 6))
names = list(accuracy_squad.keys())
values = list(accuracy_squad.values())
plt.bar(names, values, color='blue')
plt.axhline(y=accuracy_squad['Baseline LLM'], color='red', linestyle='dotted', linewidth=2, label='Baseline LLM')
plt.title('System Accuracy Comparison - SQuAD Dataset', fontsize=16)
plt.xlabel('System', fontsize=14, fontweight='bold')
plt.ylabel('Accuracy', fontsize=14, fontweight='bold')
plt.ylim(0, 1)  # Set y-axis limits to show percentages from 0 to 1
plt.grid(True, linestyle='--', linewidth=0.5, alpha=0.7)
plt.legend()
plt.show()

# QA Accuracy Plot
plt.figure(figsize=(10, 6))
names = list(qa_accuracy_squad.keys())
values = list(qa_accuracy_squad.values())
plt.bar(names, values, color='cyan')
plt.axhline(y=accuracy_squad['Baseline LLM'], color='red', linestyle='dotted', linewidth=2, label='Baseline LLM Accuracy')
plt.title('QA Accuracy for Incomplete Graph - SQuAD Dataset', fontsize=16)
plt.xlabel('Pipeline', fontsize=14, fontweight='bold')
plt.ylabel('QA Accuracy', fontsize=14, fontweight='bold')
plt.ylim(0, 1)
plt.grid(True, linestyle='--', linewidth=0.5, alpha=0.7)
plt.legend()
plt.show()

# Execution Time Plot
plt.figure(figsize=(10, 6))
names = list(execution_time_squad.keys())
values = list(execution_time_squad.values())
plt.bar(names, values, color='navy')
plt.title('Execution Time for QA - 500 Examples', fontsize=16)
plt.xlabel('Task', fontsize=14, fontweight='bold')
plt.ylabel('Execution Time (minutes)', fontsize=14, fontweight='bold')
plt.grid(True, linestyle='--', linewidth=0.5, alpha=0.7)
plt.show()

# Scalability Plot
labels = list(scalability_squad.keys())
edc_graph = [scalability_squad[label][0] for label in labels]
qa_gog = [scalability_squad[label][1] for label in labels]
qa_integrated = [scalability_squad[label][2] for label in labels]
accuracy_calc = [scalability_squad[label][3] for label in labels]

x = np.arange(len(labels))  # the label locations
width = 0.2  # the width of the bars

fig, ax = plt.subplots(figsize=(12, 8))
rects1 = ax.bar(x - width*1.5, edc_graph, width, label='EDC Graph Creation', color='lightblue')
rects2 = ax.bar(x - width/2, qa_gog, width, label='QA GoG', color='blue')
rects3 = ax.bar(x + width/2, qa_integrated, width, label='QA Integrated', color='darkblue')
rects4 = ax.bar(x + width*1.5, accuracy_calc, width, label='Accuracy Calculation', color='red')

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_xlabel('Data Sizes (Titles)', fontsize=14, fontweight='bold')
ax.set_ylabel('Execution Time (minutes)', fontsize=14, fontweight='bold')
ax.set_title('Scalability of Different Tasks', fontsize=16)
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend()
ax.grid(True, linestyle='--', linewidth=0.5, alpha=0.7)

plt.show()
```
